## df_apn
### Groupby just 'apn', 'best_stat'

```python
# as_index=False is key here. It works with .first() 
# to retreive only the first value
df_apn = df.groupby(['apn', 'best_stat'], as_index=False).first()
```

```python
df_apn.head()
```

We need to be able to get the difference in time from when the project is first filed. There are two file dates, PL FILED is first, and represents the Planning stage, then BP FILED, for the building stage. However, not all projects have both of these.

Paula Chiu from the Planning Department explained it thus in an email to me Apr 6,2017: "some projects in the development pipeline will go through review with both the Planning Department and the Department of the Building Inspection. Depending on the nature of the proposed project, an applicant may or may not need to go through review at the Planning Department and can apply for an over-the-counter permit. Most major development projects do start the intake process at the Planning Department, but there is not one linear path a project must go through from start to finish. Any applicable construction activity require permits"

```python
# pivot 
df_apn = df_apn.pivot(index='apn', columns='best_stat', values='best_date')
```

```python
df_apn.head()
```

### Create new columns for each of our dates of interest

```python
df_apn['bp_filed_to_issued_days'] = (df_apn['BP ISSUED'] - df_apn['BP FILED']).dropna().apply(lambda x: x.days)
```

```python
df_apn.head()
```

```python
df_apn['bp_issued_to_reinstated_days'] = (df_apn['BP REINSTATED'] - df_apn['BP FILED']).dropna().apply(lambda x: x.days)
```

```python
df_apn.head()
```

```python
df_apn['bp_reinstated_to_construction_days'] = (df_apn['CONSTRUCTION'] - df_apn['BP REINSTATED']).dropna().apply(lambda x: x.days)
df_apn['bp_filed_to_construction_days'] = (df_apn['CONSTRUCTION'] - df_apn['BP FILED']).dropna().apply(lambda x: x.days)
df_apn['bp_issued_to_construction_days'] = (df_apn['CONSTRUCTION'] - df_apn['BP ISSUED']).dropna().apply(lambda x: x.days)
```

```python
df_apn.head()
```
